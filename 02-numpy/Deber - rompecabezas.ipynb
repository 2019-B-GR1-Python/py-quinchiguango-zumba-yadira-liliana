```python
import numpy as np
from scipy import ndimage
from scipy import misc
import matplotlib.pyplot as plt

mapache = misc.face()

pieza_vertical1, pieza_vertical2 = np.vsplit(mapache, 2)
pieza1, pieza2 = np.hsplit(pieza_vertical1, 2)
pieza3, pieza4 = np.hsplit(pieza_vertical2, 2)
```

```python
def imprimir_piezas(pieza_superior_izq, pieza_superior_der, pieza_inferior_izq, pieza_inferior_der):
    fila1 = np.append(pieza_superior_izq,pieza_superior_der,axis=1)
    fila2 = np.append(pieza_inferior_izq,pieza_inferior_der, axis=1)
    rompecabezas=np.append(fila1,fila2,axis=0)
    plt.imshow(rompecabezas)
```

```python
def mover_piezas(pieza_mover,posicion):
    
    piezas={
            "1":pieza_sup_izq,
            "2":pieza_sup_der,
            "3":pieza_inf_izq,
            "4":pieza_inf_der,
        }

    aux = piezas[posicion]
    piezas[posicion] = piezas[pieza_mover]
    piezas[pieza_mover] = aux
    
    imprimir_piezas(piezas["1"],piezas["2"],piezas["3"],piezas["4"])
    return piezas["1"],piezas["2"],piezas["3"],piezas["4"]
```

```python
print("Rompecabezas\n")

pieza_sup_izq = pieza2
pieza_sup_der = pieza3
pieza_inf_izq = pieza4
pieza_inf_der = pieza1

imprimir_piezas(pieza_sup_izq,pieza_sup_der,pieza_inf_izq,pieza_inf_der)
plt.pause(.1)
plt.draw()

completo = 1
while completo != 0:
    pieza_mover = input("  Indique el numero de la pieza que desea mover: ")
    if (pieza_mover == 's'):
        break
    posicion = input("  Ingrese la posicion a intercambiar: ")
    pieza_sup_izq,pieza_sup_der,pieza_inf_izq,pieza_inf_der= mover_piezas(pieza_mover, posicion)
    plt.pause(.1)
    plt.draw()
```
